# Handling Missing Values: Imputation

```python
# Remove columns with all values missing
df = df.dropna(axis=1, how='all')

# Display the updated DataFrame structure to see the remaining columns
df.info()

# Remove any columns with 'Unnamed' in their name, as they typically contain no useful data
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]


# Making a for loop to loop through every column and handle its missing values by
# checking if it's numeric its gonna be changed to the mean of that column
# and if it's categorical it will replace the missing values with the mode.

# Loop through each column in the DataFrame to handle missing values
for col in df.columns:
    # Check if the column is numeric
    if pd.api.types.is_numeric_dtype(df[col]):
        # Calculate the mean of the column
        mean_value = df[col].mean()
        # Replace missing values with the mean of the column
        df[col] = df[col].fillna(mean_value)  # Assign back the filled column
        print(f"Imputed missing values in numerical column '{col}' with mean: {mean_value}")
    else:
        # For categorical columns, calculate the mode
        mode_value = df[col].mode()[0]  # [0] to get the most frequent value
        # Replace missing values with the mode of the column
        df[col] = df[col].fillna(mode_value)  # Assign back the filled column
        print(f"Imputed missing values in categorical column '{col}' with mode: {mode_value}")

# Check for any remaining missing values in the DataFrame after imputation
print("\nRemaining missing values per column after imputation:")
print(df.isnull().sum())


#Removed all the missing values and replaced them with the following mean for numeric and mode for categorical 
```

# Checking for Outliers

```python
# Identify numeric columns in the DataFrame
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Loop through each numeric column to detect outliers
for col in numeric_columns:
    # Calculate the first and third quartiles (25th and 75th percentiles)
    Q1 = df[col].quantile(0.25)  # First quartile
    Q3 = df[col].quantile(0.75)  # Third quartile
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1                # IQR is the range between Q3 and Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR  # Lower bound (anything below is an outlier)
    upper_bound = Q3 + 1.5 * IQR  # Upper bound (anything above is an outlier)
    
    # Identify rows in the DataFrame where values are outliers in this column
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    # Print the column name and calculated values for insight
    print(f"Column: {col}")
    print(f"  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")  # Show quartiles and IQR for reference
    print(f"  Lower bound: {lower_bound}, Upper bound: {upper_bound}")  # Show bounds
    print(f"  Number of outliers: {outliers.shape[0]}")  # Count of outliers in the column
    print()  # Print a blank line for readability between columns

# I was able to find the outliers and there seems to be a lot.
```

# Handling Outliers

```python
# Identify numeric columns in the DataFrame
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Loop through each numeric column to detect and replace outliers
for col in numeric_columns:
    # Calculate the first and third quartiles (25th and 75th percentiles)
    Q1 = df[col].quantile(0.25)  # First quartile
    Q3 = df[col].quantile(0.75)  # Third quartile
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1  # IQR is the range between Q3 and Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR  # Lower bound (anything below is an outlier)
    upper_bound = Q3 + 1.5 * IQR  # Upper bound (anything above is an outlier)

    # Flag to track if outliers are found in this column
    outlier_found = False
    
    # Loop through each value in the column
    for i, value in enumerate(df[col]):
        # If the value is less than the lower bound, replace it with the lower bound
        if value < lower_bound:
            df.at[i, col] = lower_bound  # Replace with lower bound
            outlier_found = True  # Set flag to True if an outlier is replaced
        # If the value is greater than the upper bound, replace it with the upper bound
        elif value > upper_bound:
            df.at[i, col] = upper_bound  # Replace with upper bound
            outlier_found = True  # Set flag to True if an outlier is replaced

    # Print results for each column
    if outlier_found:
        print(f"Outliers were found and replaced in column: {col}")
    else:
        print(f"No outliers found in column: {col}")
```
